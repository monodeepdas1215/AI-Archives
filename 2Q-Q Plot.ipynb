## Q - Q Plot

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
#std normal distribution
std_normal = np.random.normal(loc = 0, scale = 1, size=1000) # loc is the mean , scale is the std - dev
#This is how we generate random observations from a guassian distribution

#0 - 100th percentile of std-normal
for i in range(101):
    print(i, np.percentile(std_normal, i))
```

```python
#Input variable measurements is the given distribution which we are simulating here by taking another guassian distribution but in real scenario would be a given distribution
measurements = np.random.normal(loc = 20, scale = 5, size = 1000)

stats.probplot(measurements, dist='norm', plot=pylab) 
#measurements is the variable to be compared
# dist="norm" means the variable to compare with is the STANDARD NORMAL VARIATE
pylab.show()
```

```python
#generating data from a uniform distribution
measurements = np.random.uniform(low=-1, high=1, size=100000)

stats.probplot(measurements, dist='norm', plot=pylab)
pylab.show()
```

### As we can see that if measurements is not of the same type of distribution as standard normal variate then we will never be a straight line
