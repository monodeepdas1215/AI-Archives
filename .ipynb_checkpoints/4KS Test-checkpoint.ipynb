## KS TEST

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#generate a guassian  r.v X
x = stats.norm.rvs(size=1000)
sns.set_style('white')
sns.kdeplot(np.array(x), bw=0.5)
plt.show()
```

```python
#We already have a function
#here we compare it with the normal distribution
stats.kstest(x, 'norm')
```

#### Since here our p value is very high so we can say that X is also normally distributed

### Lets also see a negative case

```python
y = np.random.uniform(0, 1 ,10000)
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
#testing the above distribution
stats.kstest(y, 'norm')
```

#### As we can see that the pvalue is 0 and hence the given distribution is not normally distributed
